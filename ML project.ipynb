```python
non_numeric_columns=data.select_dtypes(exclude=['number']).columns
print("Non numeric columns : ",non_numeric_columns)
```

```python
label_encoder=LabelEncoder()
```

```python
for i in non_numeric_columns:
    data[i]=label_encoder.fit_transform(data[i])
```

```python
data.head()
```

```python
variance=data.var()
variance = variance.reset_index()
variance.columns = ['ID', 'values']
variance = variance.sort_values('values', ascending=1)
print(variance)
```

```python
lessvar=variance.loc[variance['values']<0.2,'ID']
```

```python
lessvar
```

```python
data1=data.drop(lessvar,axis=1)
data1.head()
```

```python
data1[data1.isnull().any(axis=1)]
```

```python
cl = data1.corr().abs()
ts = cl.unstack()
ts = pd.DataFrame(ts)
```

```python
ts.reset_index(inplace=True)
ts.head()
```

```python
ts['flag'] = np.where(ts['level_0']==ts['level_1'],'same','not same')
ts.columns.values[2]='corr'
ts.head()
```

```python
name = ts.loc[(ts['corr']>0.9) & (ts['flag']!='same'),'level_1']
final_features=name.unique()
final_features
```
